```python
import seaborn as sns
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

results = pd.read_csv("Experiment results - Sheet1.csv")
results
```

```python
results.isna().any(axis =1)
```

```python
results.dropna(inplace = True)
results
```

```python
attempt = pd.read_csv("Experiment results - Attempt.csv")
attempt.dropna(inplace = True)
attempt 
#this dataset combines the columns: "Location change in length" and "Location change 
#in population" in the results dataset into one ("Location change").
#Done to use for the hue parameter
```

```python
#Using how many changes as x
```

```python
# Draw a pointplot to show pulse as a function of three categorical factors
sns.set_theme(style="whitegrid")

g = sns.catplot(
    data= attempt, x="how many changes", y="Power", hue="Location change", col="Change",
    capsize=.2, palette="YlGnBu_d", errorbar="ci",
    kind="point", height=6, aspect=.75,
)
g.despine(left=True)
```
