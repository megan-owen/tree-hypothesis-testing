```python
import seaborn as sns
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

results = pd.read_csv("experiment-results.csv")
results
```

```python
empty = results.isna().any(axis =1)
results = results[~empty]
results
```

```python
attempt = pd.read_csv("experiment-results - Attempt.csv")

empty = attempt.isna().any(axis =1)
attempt = attempt[~empty]

attempt 
#this dataset combines the columns: "Location change in length" and "Location change 
#in population" in the results dataset into one ("Location change").
#Done to use for the hue parameter
```

```python
#Using how many changes as x
```

```python
# Draw a pointplot to show pulse as a function of three categorical factors
sns.set_theme(style="whitegrid")

g = sns.catplot(
    data= attempt, x="how many changes", y="Power", hue="Location change", col="Change",
    capsize=.2, palette="YlGnBu_d", errorbar="ci",
    kind="point", height=6, aspect=.75,
)
g.despine(left=True)
```

```python
#For tree comparisons done with this three as the first tree and at most one change:
#gts_dendropy_CAT_(((AB)C)D)_tauAB-10000_tauABC-20000_tauRoot-30000_pAB-10000_pABC-10000_pRoot-10000

filter1= results["tau_AB1"]==10000
filter2= results["tau_ABC1"]==20000
filter3= results["tau_Root1"]==30000
filter4= results["p_AB1"]==10000
filter5= results["p_ABC1"]==10000
filter6= results["p_Root1"]==10000
filter7= results["how many changes"]<=1 

results1= results[filter1&filter2&filter3&filter4&filter5&filter6&filter7]
results1
```
