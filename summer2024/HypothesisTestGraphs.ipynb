```python
import seaborn as sns
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv("experiment-results.csv")
results
```

```python
empty = results.isna().any(axis =1)
results = results[~empty]
results
```

```python
results["Power"] = results["Power"].astype(float)
```

```python
#For tree comparisons done with this three as the first tree and at most one change:
#gts_dendropy_CAT_(((AB)C)D)_tauAB-10000_tauABC-20000_tauRoot-30000_pAB-10000_pABC-10000_pRoot-10000

filter1= results["tau_AB1"]==10000
filter2= results["tau_ABC1"]==20000
filter3= results["tau_Root1"]==30000
filter4= results["p_AB1"]==10000
filter5= results["p_ABC1"]==10000
filter6= results["p_Root1"]==10000
filter7= results["how many changes"]<=1 

results1= results[filter1&filter2&filter3&filter4&filter5&filter6&filter7]
results1
```

```python
sns.set_theme(style="whitegrid")

g = sns.catplot(
    data= results1, x="tau_AB2", y="Power", hue="Change",
    capsize=.2, palette="YlGnBu_d", errorbar= ("ci",95),
    kind="point", height=4, aspect=1,
)
g.despine(left=True)
```
