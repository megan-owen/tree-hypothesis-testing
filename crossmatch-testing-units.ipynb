<h3> Crossmatch funtion testing unit</h3>

```python

#testing crossmatch function 

#This function will create a random matrix to then use the python and R functions to compare the outputs
# run_num is the number of the current run used to differentiate runs for same n_size matrix
# n_size is the size of the NxN random symetric matrix desired

import rpy2.robjects as robjects
import pandas as pd
import rpy2.robjects.numpy2ri
import numpy as np
import os
import time
import functools
import json
import networkx as nx
rpy2.robjects.numpy2ri.activate()

#testing symetric matrix to replace for matrix functions
def testing_python_R_functions(run_num,n_size):
    
    N = n_size
    b = np.random.randint(1,150,size=(N,N))
    # b = np.random.rand(N, N)
    b_symm = (b + b.T)/2
    for i in range(0,N,1):
        for j in range(0,N,1):
            if(i == j):
                b_symm[i][j] = 0
    # print("Python matrix: ",b_symm)

    # b_symm = pd.read_csv("C:/Users/Eric Aragundi/Desktop/b_symm3.csv",header=None).to_numpy()
    # print("Python matrix: ",b_symm)
    b_symm_to_save = pd.DataFrame(b_symm)
    
    crossmatch_trials_output = data_dir + 'crossmatch trials/'
            
    if not os.path.exists(crossmatch_trials_output):

      # Create a new directory because it does not exist 
      os.makedirs(crossmatch_trials_output)
      print("The crossmatch trials output directory is created!")
    
    
    b_symm_to_save.to_csv(f"{crossmatch_trials_output}b_symm_{n_size}_{run_num}.csv",header=False,index=False)
    print("file location: "+ f"b_symm_{n_size}_{run_num}")
    
    ##this are not random matrices that can be used as well for testing
#     x = np.array([0, 1])
#     z = np.array([[1,2,3,2],[2,1,6,5],[3,6,1,4],[2,5,4,1]])
#     z2 = np.array([[0,1,1,2],[1,0,2,3],[1,2,0,4],[2,3,4,0]])
#     vec_z = np.repeat(x, [len(z)/2], axis=0)
#     vec_z2 = np.repeat(x, [len(z2)/2], axis=0)
    
    
    vec2 = np.repeat(x, [len(b_symm)/2], axis=0)
    #print(vec2)
    python_res_list = cf.crossmatchtest(vec2,b_symm)
    print("python function output:\n",python_res_list)



    robjects.r('''
            # create a function `f`
            f <- function(df1) {
                library(crossmatch)
                data <-  data.matrix(df1)
                data
                z <- c(rep(0,(nrow(data))/2),rep(1,(nrow(data))/2)) # creates vector

                answer<- crossmatch::crossmatchtest(z,data)

            }
            # call the function `f` with argument value 3
           #f(data)
            ''')
    r_f = robjects.globalenv['f']

    res = r_f(b_symm)
    cross_test = dict(zip(res.names, map(list,list(res))))
    final_cross_list = {
                        'a1':cross_test['a1'][0],
                        'Ea1':cross_test['Ea1'][0],
                        'Va1':cross_test['Va1'][0],
                        'dev':cross_test['dev'][0],
                        'pval':cross_test['pval'][0],
                        'approxval':cross_test['approxpval'][0]
                     }
    print("R function output:\n",final_cross_list)
    
    comparison_result = ""
            
    if functools.reduce(lambda x, y : x and y, map(lambda p, q: abs(p - q) < .0000001,python_res_list.values(),final_cross_list.values()), True): 
        comparison_result = "\nThe lists python list and R list are the same\n"
        print (comparison_result) 
    else: 
        comparison_result = "\nThe lists python list and R list are not the same\n"
        print (comparison_result) 
        
    with open(f'C:/Users/Eric Aragundi/Desktop/crossmatch trials/b_symm_{n_size}_{run_num}.txt', 'w') as f:
        python_res_string = json.dumps(python_res_list)
        R_res_string = json.dumps(final_cross_list)
        final_line = "python function output:\n" + python_res_string + "R function output:\n" + R_res_string + comparison_result
        f.write(final_line)

```

```python

# testing all symetric matrix from size 4 up to size 100 going up by 4. We are doing 10 test for each n_size

for n_size in range(4,101,4):
    for run_num in range(1,11,1): 
    #     run_num = 2
    #     n_size = 100
        testing_python_R_functions(run_num,n_size)
```
